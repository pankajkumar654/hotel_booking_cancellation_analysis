```python
df = df[df['adr']<5000]
```

 ## analysis and visualization   

```python
cancel_percentage = df['is_canceled'].value_counts(normalize = True)
cancel_percentage

plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python

# converting both columns in string
df['hotel'] = df['hotel'].astype(str)
df['is_canceled'] = df['is_canceled'].astype(str)

plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Get the legend labels
legend_labels, _ = ax1.get_legend_handles_labels()
# Set legend with labels
ax1.legend(legend_labels, ['Not canceled', 'Canceled'], bbox_to_anchor=(1, 1))

ax1.set_title('Reservation status in different hotels', size=20)
ax1.set_xlabel('Hotel')
ax1.set_ylabel('Number of reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)

# in resort hotel 27% reservations are getting cancelled 
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)

# in city hotels 41% resversations are getting cancelled
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(18,8))
plt.title('Average daily rate in city and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label= 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label= 'City Hotel')
plt.legend(fontsize = 20)
plt.show()

# we can see in  weekends and holidays the resort hotel rates spikes  
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x= 'month', hue='is_canceled',data = df)

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not canceled', 'Canceled'], bbox_to_anchor=(1, 1))
ax1.set_title('Reservation status per month', size=20)
ax1.set_xlabel('month')
ax1.set_ylabel('no of resevations')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Adr per month', fontsize = 30)
data =  df[df['is_canceled'] == '1'].groupby('month')['adr'].sum().reset_index()
sns.barplot(x='month',y='adr', data=data)
plt.show()
```

```python
cancelled_data = df[df['is_canceled']=='1']
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels = top_10_country.index)
plt.show()
```

```python
cancelled_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace = True)
cancelled_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data = df[df['is_canceled']=='0']
not_cancelled_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace = True)
not_cancelled_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily rate')
plt.plot(not_cancelled_adr['reservation_status_date'],not_cancelled_adr['adr'],label = 'not cancelled')

plt.plot(cancelled_adr['reservation_status_date'],cancelled_adr['adr'],label = 'cancelled')
plt.legend()
plt.show()
```

```python
cancelled_adr = cancelled_adr[(cancelled_adr['reservation_status_date']>'2016') & (cancelled_adr['reservation_status_date']<'2017-09')]

not_cancelled_adr = not_cancelled_adr[(not_cancelled_adr['reservation_status_date']>'2016') & (not_cancelled_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize = 40)
plt.plot(not_cancelled_adr['reservation_status_date'],not_cancelled_adr['adr'], label='not cancelled')
plt.plot(cancelled_adr['reservation_status_date'],cancelled_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()


# we can see in the graph, reservations are canceled when the average daily rate is higher than when it is not canceled.
# It clearly proves all the above analysis that the higher price leads to higher cancellation
```
